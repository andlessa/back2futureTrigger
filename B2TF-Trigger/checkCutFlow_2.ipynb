```python
import pandas as pd
import numpy as np
pd.options.display.float_format = '{:.5g}'.format
```

```python
file = './benchmark500_delphes_events_b2tf_cutFlow.csv'

cutDF = pd.read_csv(file,skiprows=2,delimiter=';')
if len(cutDF) > 1:
    cutDF = cutDF[abs(cutDF['ctau']-9.0) < 0.5]
newRows = {}
for key in cutDF.columns:
    newRows[key] = [np.sqrt(float(cutDF[key].iloc[0])),
                     float(cutDF[key].iloc[0]/cutDF['# All'].iloc[0])]

df_the_dict = pd.DataFrame.from_dict(newRows)

cutDF2 = pd.concat([cutDF, df_the_dict], ignore_index=True)
cutDF2.drop(columns=['tau'],inplace=True)
cutDF2.rename({0 : 'entries', 1 : 'entriesError', 2 : 'cumulative'},axis=0,inplace=True)
# for col in cutDF2.columns:
#     cutDF2[col] = cutDF2[col]*40000.0/cutDF2['# All'].iloc[0]
cutDF2.T
```
